## Data Cleaning

```python
#menghapus kolom yang tidak dibutuhkan
df1 = df.drop(['beauty_point_earned', 
               'price_by_combinations', 
               'url', 'active_date', 
               'rating_types_str', 
               'average_rating_by_types', 
               'total_repurchase_maybe_count', 
               'total_in_wishlist',
               ], axis=1)
```

```python
df1
```

```python
#memisahkan nomor id pada brand dengan nama brand dari kolom brand_name
df1[['brand_id', 'brand_name']] = df1['brand_name'].str.split('_', n = 1, expand = True)
#memisahkan range harga dari kolom price_range
df1[['price_range', 'price_range2']] = df1['price_range'].str.split('-', n = 1, expand = True)
#memisahkan categori produk dari kolom categories
df1[['category', 'category2', 'subcategory']] = df1['categories'].str.split(';', n = 2, expand = True)

```

```python
#mengurutkan kolom
df1 = df1[['brand_id', 
           'brand_name', 
           'product_name',
           'product_id', 
           'price_range',
           'price_range2',
           'category',
           'subcategory',
           'average_rating',
           'total_reviews',
           'total_recommended_count',
           'total_repurchase_no_count', 
           'total_repurchase_yes_count',
           ]]
```

```python
df1.columns
```

```python
#menghapus row yang tidak dipakai 
df1 = df1[df1.apply(lambda row: row.astype(str).str.contains('None').any(), axis=1)]
del df1['price_range2']

df1 = df1[(df1['subcategory'] == None) | (df1['average_rating'] != 0.000000) & (df1['total_reviews'] != 0)].dropna()
df1
```

```python
#menghilangkan Rp dan (.) pada kolom price
df1['price_range'] = df1['price_range'].str.replace('Rp', '')
df1['price_range'] = df1['price_range'].str.replace('.', '')

#mengganti nama kolom
df1.rename(columns = {'price_range':'price'}, inplace = True) 
df1.rename(columns = {'total_reviews':'amount_sold'}, inplace = True) 
```

# D. Analisis dan Perhitungan

1. Bagaimana daya beli calon konsumen Sociolla berdasarkan kategori produk?

```python
purchase_power = []

for i in df1['price']:
    if i <= 100000:
        purchase_power.append('low')
    elif i <= 500000:
        purchase_power.append('medium')
    else:
        purchase_power.append('high')

df1['purchase_power'] = purchase_power
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x= 'purchase_power', hue='category', data= (df1[df1['purchase_power'] == 'low']))
plt.title('Daya Beli Konsumen Berdasarkan Kategori Produk')
plt.xlabel('Daya Beli')
plt.ylabel('Jumlah Pembelian')
plt.legend(title='Kategori Produk')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x= 'purchase_power', hue='category', data= (df1[df1['purchase_power'] == 'medium']))
plt.title('Daya Beli Konsumen Berdasarkan Kategori Produk')
plt.xlabel('Daya Beli')
plt.ylabel('Jumlah Pembelian')
plt.legend(title='Kategori Produk')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x= 'purchase_power', hue='category', data= (df1[df1['purchase_power'] == 'high']))
plt.title('Daya Beli Konsumen Berdasarkan Kategori Produk')
plt.xlabel('Daya Beli')
plt.ylabel('Jumlah Pembelian')
plt.legend(title='Kategori Produk')
plt.show()
```

Dari data di atas, dapat disimpulkan bahwa dari seluruh tingkat daya beli konsumen Sociolla, kategori produk yang memiliki jumlah pembelian tertinggi adalah kategori skincare dibandingkan produk yang lain.

2. Apa brand yang paling diminati konsumen?

```python
minat = df1.groupby(['brand_name'], as_index=False).agg({'amount_sold': 'sum', 'average_rating' : 'mean'})

# Ambil data top 5 produk
topminat = minat.sort_values('amount_sold', ascending=False).head(10)

plt.figure(figsize=(8, 6))
sns.barplot(x= 'brand_name', y = 'amount_sold', hue = 'amount_sold', data = topminat)
plt.xticks(rotation=45)
plt.title('Top 10 Brand yang Paling Diminati Konsumen')
plt.xlabel('Nama Brand')
plt.ylabel('Total Pembelian')
plt.show()
```

Berdasarkan data di atas, 10 nama brand yang paling diminati oleh konsumen Sociolla adalah avoskin, azarine, ariul, biore, barenbliss, blp-beauty, benton, bioderma, banana-boat, dan acwell dengan berdasarkan dari banyaknya produk terjual. 

3. Kategori produk apa yang memiliki pendapatan yang paling tinggi?

```python
# Menghitung total pendapatan untuk setiap kategori produk
df1['revenue'] = df1['price']*df1['amount_sold']
```

```python
income = df1.groupby(['subcategory', 'category'], as_index=False)['revenue'].sum()

# Ambil data top 10 kategori produk dengan pendapatan tertinggi
top = income.sort_values('revenue', ascending=False).head(10)

# Plot barplot untuk memvisualisasikan kategori produk dengan pendapatan tertinggi
plt.figure(figsize=(10, 6))
sns.barplot(x='subcategory', y='revenue', hue='category', data=top)
plt.title('Top 10 Kategori Produk dengan Pendapatan Tertinggi')
plt.xlabel('Kategori Produk')
plt.ylabel('Total Pendapatan')
plt.xticks(rotation=90)
plt.show()
```

Dari data di atas, dapat diketahui bahwa kategori skincare dengan jenis produknya yaitu face serum menjadi produk dengan pendapatan tertinggi. Selain itu, dilihat dari warna bar pada jenis jenis produk yang lain, kategori skincare di beberapa jenis produk mendominasi dalam 10 urutan pendapatan tertinggi dibandingkan kategori produk yang lain. 

4. Apakah harga mempengaruhi kepuasan konsumen?

```python
# Plot scatter plot untuk melihat hubungan antara harga dan rating
plt.figure(figsize=(8, 6))
sns.scatterplot(x='price', y='average_rating', hue='category', data=df1[df1['average_rating'] > 4.5])
plt.title('Hubungan Antara Harga dan Rating Produk')
plt.xlabel('Harga')
plt.ylabel('Rating Rata-rata')
plt.show()
```

Dari data di atas, dapat diketahui bahwa lebih banyak produk murah di sociolla yang didominasi oleh kategori skincare, rata-rata rating dengan harga produk di bawah Rp 400.000 adalah 4.5 sampai dengan 5. Kemudian, tidak terdapat hubungan antara harga dan rating produk di Sociolla. 

5. Apakah kepuasan konsumen mempengaruhi keputusan untuk membeli kembali produk?

H0 = Kepuasan konsumen tidak mempengaruhi keputusan konsumen untuk membeli kembali produk Sociolla

H1 = Kepuasan konsumen mempengaruhi keputusan konsumen untuk membeli kembali produk Sociolla

```python
from scipy import stats

#menemukan korelasi dengan uji Spearman
corr_rho, pval_s = stats.spearmanr(df1['average_rating'], df1['total_repurchase_yes_count'])
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")
```

Dalam hal ini, nilai korelasi Spearman sebesar 0.31 menunjukkan adanya hubungan negatif antara rating produk dan total keputusan untuk membeli kembali. nilai korelasi negatif menunjukkan hubungan yang lemah. Nilai p-value sangat kecil yaitu sebesar 4.1019553715704115e-20, yang menunjukkan bahwa korelasi antara kepuasan konsumen dan keputusan untuk membeli kembali adalah sangat signifikan sehingga H1 diterima.

6. Apakah terdapat perbedaan rata rata pembelian pada produk kategori makeup dan skincare?

```python
# Filter DataFrame untuk kategori Makeup dan Skincare
makeup_data = df1[df1['category'] == 'Makeup']['amount_sold']
skincare_data = df1[df1['category'] == 'Skincare']['amount_sold']

# Lakukan uji independent t-test
t_stat, p_val = stats.ttest_ind(makeup_data, skincare_data)

print('T-Statistic:',t_stat)
print('P-value:',p_val)
```

Nilai statistik negatif -0.6528318331979693 dan P-value sebesar 0.5141825359652853 menunjukkan bahwa perbedaan tidak signifikan secara statistik pada tingkat kepercayaan 95%. Dengan kata lain, tidak terdapat perbedaan yang signifikan antara rata-rata pembelian produk Makeup dan Skincare.
